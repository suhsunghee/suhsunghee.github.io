```python
import pandas as pd

df = pd.read_csv('Absenteeism_data.csv')
```

```python
df
```

```python
df = df.drop(['ID'], axis = 1)
```

```python
## overviewing reason for absence

sorted(df['Reason for Absence'].unique())
```

```python
#creating dummy column for Reason for Absence

reason_columns = pd.get_dummies(df['Reason for Absence'])
```

```python
reason_columns
```

```python
#creating check column to make sure there is only one '1' for each index
reason_columns['check'] = reason_columns.sum(axis=1)
```

```python
reason_columns.head()
```

```python
reason_columns['check'].unique()
```

```python
reason_columns['check'].sum(axis = 0)
```

```python
reason_columns = pd.get_dummies(df['Reason for Absence'], drop_first = True)
```

```python
#removing reason for absence from original dataframe
df = df.drop(['Reason for Absence'], axis = 1 )
```

```python
#Grouping  dummy variables into categorical variables
#Reasons for absence are grouped into 4 categories

reason_type_1 = reason_columns.loc[:,1:14].max(axis =1)
reason_type_2 = reason_columns.loc[:,15:17].max(axis =1)
reason_type_3 = reason_columns.loc[:,18:21].max(axis =1)
reason_type_4 = reason_columns.loc[:,22:].max(axis =1)
```

```python
#concatenate reasons and original dataframe

df = pd.concat([df, reason_type_1,reason_type_2,reason_type_3,reason_type_4], axis = 1)
```

```python
df
```

```python
# Renaming last four reason columns and sorting it 
df.columns.values
```

```python

column_new_names = ['Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours','Reason_1','Reason_2', 'Reason_3', 'Reason_4']
```

```python
df.columns = column_new_names
```

```python
df.head(3)
```

```python
column_name_sort =['Reason_1','Reason_2', 'Reason_3', 'Reason_4','Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']

df = df[column_name_sort]
```

```python
df.head()
```

```python
#Creating Checkpoint

df_reason_mod = df.copy()
```

```python
#Date formating to extract month value and creating 'Month'column in original dataframe
type(df_reason_mod['Date'])
```

```python
df_reason_mod['Date']= pd.to_datetime(df_reason_mod["Date"], format = '%d/%m/%Y')
```

```python
df_reason_mod.shape
```

```python
list_months = []
for i in range (df_reason_mod.shape[0]:
    list_months.append(df_reason_mod['Date'][i].month)
```

```python
df_reason_mod['Month'] = list_months
```

```python
df_reason_mod
```

```python
#adding a weekday value

def date_to_weekday(date_value):
    return date_value.weekday()

df_reason_mod['Day of the Week'] = df_reason_mod['Date'].apply(date_to_weekday) 
```

```python
df_reason_mod.head()
```

```python
# Analyzing and transforming Education data
#Grouping 1 - highschool graduate 2- graduate 3-post grad 4 - Phd into 2 categories 1->0 2,3,4->1

df_reason_mod['Education'].unique()
```

```python
df_reason_mod['Education'] = df_reason_mod['Education'].map({1:0,2:1,3:1,4:1})
```

```python
df_reason_mod['Education'].value_counts()
```

```python
df_preprocessed = df_reason_mod.copy()
```
