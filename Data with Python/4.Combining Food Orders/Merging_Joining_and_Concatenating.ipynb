```python
# Merging Joining and Concatenating 

import pandas as pd
week1 = pd.read_csv("Restaurant - Week 1 Sales.csv")
week2 = pd.read_csv("Restaurant - Week 2 Sales.csv")
customers = pd.read_csv("Restaurant - Customers.csv")
foods = pd.read_csv("Restaurant - Foods.csv")
```

```python
# overview of weekly data 
week1.head(3)
```

```python
week2.head(3)
```

```python
sales = pd.concat(objs = [week1, week2], keys = ["Week 1","Week 2"])
sales.head(3)
```

```python
## Performing inner joins 
week1.merge(week2, how = "inner", on = "Customer ID", suffixes = ["_Week1", "_Week2"])
```

```python
# multiple common fields inner join 

week1.merge(week2, how = "inner", on = ["Customer ID","Food ID"])
```

```python
# outer join and turning on the indicator to remove rows that exist in both week1 and week2 

merged = week1.merge(week2, how = "outer", on = "Customer ID", suffixes = ["_Week1","_Week2"], indicator = True )
```

```python
# from this, we can see there are 62 customer who came in on week1 and week2 

merged["_merge"].value_counts()
```

```python
# using filters to see customer who only came in either week 1 or week 2 
mask = merged["_merge"].isin(["left_only","right_only"])
merged[mask]
```

```python
# left join on columns with different names and dropping the duplicated columns

week2.merge(customers, how = "left", left_on ="Customer ID", right_on= "ID").drop("ID", axis ="columns" )
```
